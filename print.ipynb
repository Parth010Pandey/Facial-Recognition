```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = pd.read_csv('ck log1.csv')
x.head()
# x(0)
```

```python
def mapFun(x):
	i = x.find('Training Acuuarcy ')
	n = x[i+18:i+24]
	n = float(n)
	return n
training =  x['data']
training = training.map(mapFun)
```

```python
def mapFun(x):
	i = x.find('Validation Acuuarcy ')
	n = x[i+20:i+26]
	n = float(n)
	return n
testing = x['data']
testing =  testing.map(mapFun)
```

```python
plt.figure()
plt.xlabel('Epochs')
plt.ylabel("Accuracy")
plt.title("Accuracy on CK+")

plt.plot(training, label='Training')
plt.plot(testing, label='Testing')
plt.legend()
```

```python
def mapFun(x):
	i = x.find('Training Loss: ')
	n = x[i+15:i+21]
	n = float(n)
	return n
trainingLoss =  x['data']
trainingLoss = trainingLoss.map(mapFun)
```

```python
def mapFun(x):
	i = x.find('Validation Loss ')
	n = x[i+16:i+22]
	n = float(n)
	return n
testingLoss =  x['data']
testingLoss = testingLoss.map(mapFun)
```

```python
plt.figure()
plt.xlabel('Epochs')
plt.ylabel("Loss")
plt.title("Loss on CK+")

plt.plot(trainingLoss, label='Training')
plt.plot(testingLoss, label='Testing')
plt.legend()
```
